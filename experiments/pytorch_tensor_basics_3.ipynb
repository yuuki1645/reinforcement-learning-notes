# PyTorch テンソル連結入門：`torch.cat` 徹底理解

**目的**: 複数のテンソルを、指定した次元 (`dim`) に沿って「くっつける（連結する）」関数。

- **書式**: `torch.cat(tensors, dim=0)`
- **引数**:
  - **`tensors`**: テンソルのリストやタプル（例: `[t1, t2, t3]`）
  - **`dim`**: どの次元方向に連結するか（0 が行方向、1 が列方向、など）
- **条件**:
  - `dim` 以外の次元サイズは、すべてのテンソルで一致している必要があります。

### `cat` という名前の由来

`cat` は英単語の「猫」ではなく、**concatenate（連結する）** の略です。

- **concatenate**: 「つなげる」「連結する」という意味の英単語
- `torch.cat` = 「テンソルを連結する関数」という名前

同様に、NumPy の `np.concatenate` も同じ語源で、配列を連結する関数です。

### よく出るイメージ

2 次元テンソル（行列）を例にします。

- **`dim=0`**: 行方向に縦につなげる（行が増える）
- **`dim=1`**: 列方向に横につなげる（列が増える）

```python
import torch

# 基本例: 1 次元テンソルを連結
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])

print("x =", x)
print("y =", y)

z = torch.cat([x, y], dim=0)
print("torch.cat([x, y], dim=0) =", z)
print("shape:", z.shape)  # → torch.Size([6])
```

```python
import torch

# 2 次元テンソルの例（行・列方向の連結）
A = torch.tensor([[1, 2],
                  [3, 4]])
B = torch.tensor([[5, 6],
                  [7, 8]])

print("A =\n", A)
print("B =\n", B)

# dim=0: 行方向（縦）に連結 → 行数が増える
C0 = torch.cat([A, B], dim=0)
print("\ntorch.cat([A, B], dim=0) =\n", C0)
print("shape:", C0.shape)  # → torch.Size([4, 2])

# dim=1: 列方向（横）に連結 → 列数が増える
C1 = torch.cat([A, B], dim=1)
print("\ntorch.cat([A, B], dim=1) =\n", C1)
print("shape:", C1.shape)  # → torch.Size([2, 4])
```

```python
import torch

# 3 つ以上のテンソルを連結する例
A = torch.tensor([[1, 2],
                  [3, 4]])
B = torch.tensor([[5, 6],
                  [7, 8]])
D = torch.tensor([[ 9, 10],
                  [11, 12]])

C_multi = torch.cat([A, B, D], dim=0)
print("3 つ以上のテンソルを連結 (dim=0) =\n", C_multi)
print("shape:", C_multi.shape)
```

### `torch.cat` と `torch.stack` の違い（イメージ）

- **`torch.cat`**
  - 既にある次元を「伸ばして」くっつける
  - **次元数は増えない**（1 次元→1 次元のまま、2 次元→2 次元のまま）
- **`torch.stack`**
  - テンソルを「積み重ねて」、新しい次元を 1 つ増やす
  - **次元数が 1 つ増える**（1 次元→2 次元、2 次元→3 次元 など）

#### 直感的なイメージ

- `cat`: 「列をつなげる」「行をつなげる」など、**横や縦に並べて 1 本につなぐ**
- `stack`: 「紙を重ねる」「画像を束ねる」ように、**同じ形のテンソルを積み重ねて束にする**

下のコードセルで、1 次元テンソルの例と 2 次元テンソルの例を順番に確認します。

```python
import torch

# 1 次元テンソルでの違い
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])

print("x:", x, "shape:", x.shape)  # torch.Size([3])
print("y:", y, "shape:", y.shape)  # torch.Size([3])

# --- torch.cat: 1 次元をそのまま伸ばす（次元数は増えない） ---
cat_1d = torch.cat([x, y], dim=0)
print("\ncat 結果:", cat_1d)
print("cat shape:", cat_1d.shape)  # torch.Size([6])

# --- torch.stack: 新しい次元を追加して 2 次元にする ---
stack_1d_0 = torch.stack([x, y], dim=0)  # 0 次元目に積む
print("\nstack(dim=0) 結果:\n", stack_1d_0)
print("stack(dim=0) shape:", stack_1d_0.shape)  # torch.Size([2, 3])

stack_1d_1 = torch.stack([x, y], dim=1)  # 1 次元目に積む
print("\nstack(dim=1) 結果:\n", stack_1d_1)
print("stack(dim=1) shape:", stack_1d_1.shape)  # torch.Size([3, 2])
```

```python
import torch

# 2 次元テンソルでの違い
A = torch.tensor([[1, 2],
                  [3, 4]])
B = torch.tensor([[5, 6],
                  [7, 8]])

print("A shape:", A.shape)  # torch.Size([2, 2])
print("B shape:", B.shape)  # torch.Size([2, 2])

# --- torch.cat: 行方向に連結（行数が増えるだけ、2 次元のまま） ---
cat_2d = torch.cat([A, B], dim=0)
print("\ncat(dim=0) =\n", cat_2d)
print("cat(dim=0) shape:", cat_2d.shape)  # torch.Size([4, 2])

# --- torch.stack: 「2 枚の行列の束」にする（新しい次元が増えて 3 次元） ---
stack_2d = torch.stack([A, B], dim=0)
print("\nstack(dim=0) =\n", stack_2d)
print("stack(dim=0) shape:", stack_2d.shape)  # torch.Size([2, 2, 2])
```

```python
import torch

# torch.stack との違いを確認
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])

S = torch.stack([x, y], dim=0)
print("stack の例: torch.stack([x, y], dim=0) =\n", S)
print("shape:", S.shape)  # → torch.Size([2, 3])  新しい次元が増える
```
