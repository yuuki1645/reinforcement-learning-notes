```python
import torch
import torch.nn as nn

# 入力3次元 → 中間4次元 → 出力2次元の2層全結合モデル
layer1 = nn.Linear(3, 4)
layer2 = nn.Linear(4, 2)

# まとめて書きたい場合は Sequential でもOK
model = nn.Sequential(
    layer1,
    nn.ReLU(),  # 活性化を挟む（不要なら消してOK）
    layer2,
)

model
```

```python
# ダミー入力（バッチサイズ=4, 特徴量=3）をモデルに通してみる
x = torch.randn(4, 3)
with torch.no_grad():
    y = model(x)

print("input shape:", x.shape)
print("output shape:", y.shape)
y
```

```python
x_1d = torch.randn(3)
with torch.no_grad():
  y_1d = model(x_1d)
y_1d
```
